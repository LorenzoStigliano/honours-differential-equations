```python
import sympy as sym
import sympy.plotting as sym_plot
from sympy import sin
sym.init_printing()
from IPython.display import display_latex

# Define the equation
x = sym.symbols('x')
y = sym.Function('y')
eq1 = sym.Eq(y(x).diff(x), -y(x)**2*sin(x))
print("The equation")
print(eq1)
display_latex(eq1)

eq1sol = sym.dsolve(eq1, y(x), ics={y(0):1})
print("has solutions")
display_latex(eq1sol)
print("or equivalently")
display_latex(sym.simplify(eq1sol))

a = sym_plot.plot(eq1sol.rhs, (x,0,10), xlabel = 'x', ylabel = 'y')
```

```python
t = sym.symbols('t')
x1 = sym.Function('x1')
x2 = sym.Function('x2')
system = [sym.Eq(x1(t).diff(t), 2*x1(t)),
          sym.Eq(x2(t).diff(t), 5*x1(t)-3*x2(t))]
print("The system")
display_latex(system)
system_sol = sym.dsolve(system)
print("has solution")
display_latex(system_sol)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the equation
def dX_dt(X, t):
    x, y = X
    return [y, x-x**3]

# Set up the time samples
t = np.linspace(0, 10, 100)

for y in np.arange(0,3,0.5):
    # Initial conditions
    X0 = [0, y]
    
    # Solve
    Xsol = odeint(dX_dt, X0, t)
    
    # Plot the (x1, x2) coordinates of the solution
    plt.plot(Xsol[:, 0], Xsol[:, 1])

plt.legend(loc='best')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid()
```
